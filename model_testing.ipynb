## Define Training and Test Sets

```python
# Define x and y
X_combined_feats = df_final.drop(['Recommend', 'Date_Epoch_Week'], axis=1).values
X_keyword_only_feats = df_final.drop(['Num_Helpful', 'Num_Funny', 'Hours_Played'], axis=1).values
X_nonkeyword_feats = df_final[['Num_Helpful', 'Num_Funny', 'Hours_Played']].values
#X = df_final.drop(['Recommend'], axis=1).values
y = df_final['Recommend'].values
```

```python
print(X_combined_feats)
print(X_combined_feats.shape)
```

```python
print(X_keyword_only_feats)
print(X_keyword_only_feats.shape)
```

```python
print(X_nonkeyword_feats)
print(X_nonkeyword_feats.shape)
```

```python
# Split dataset into Train and Test for 3 categories: 1. combined features; 2. keyword only features; 3. nonkeyword features
from sklearn.model_selection import train_test_split
X_train_combined_feats, X_test_combined_feats, y_train_combined_feats, y_test_combined_feats = train_test_split(X_combined_feats, y, test_size=0.2, random_state=0)
X_train_keyword_only_feats, X_test_keyword_only_feats, y_train_keyword_only_feats, y_test_keyword_only_feats = train_test_split(X_keyword_only_feats, y, test_size=0.2, random_state=0)
X_train_nonkeyword_feats, X_test_nonkeyword_feats, y_train_nonkeyword_feats, y_test_nonkeyword_feats = train_test_split(X_nonkeyword_feats, y, test_size=0.2, random_state=0)
```

```python
print(X_train_combined_feats.shape)
print(y_train_combined_feats.shape)
print(X_test_combined_feats.shape)
print(y_test_combined_feats.shape)
print()

print(X_train_keyword_only_feats.shape)
print(y_train_keyword_only_feats.shape)
print(X_test_keyword_only_feats.shape)
print(y_test_keyword_only_feats.shape)
print()

print(X_train_nonkeyword_feats.shape)
print(y_train_nonkeyword_feats.shape)
print(X_test_nonkeyword_feats.shape)
print(y_test_nonkeyword_feats.shape)
```

```python
# Normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_combined_feats = scaler.fit_transform(X_train_combined_feats)
X_test_combined_feats = scaler.transform(X_test_combined_feats)

X_train_keyword_only_feats = scaler.fit_transform(X_train_keyword_only_feats)
X_test_keyword_only_feats = scaler.transform(X_test_keyword_only_feats)

X_train_nonkeyword_feats = scaler.fit_transform(X_train_nonkeyword_feats)
X_test_nonkeyword_feats = scaler.transform(X_test_nonkeyword_feats)
```

```python
# # Just to make sure values are normalized
# X_train[0:15]
```

# Multilinear Regression

```python
# Train model on training sets
from sklearn.linear_model import LinearRegression
regressor1 = LinearRegression().fit(X_train_combined_feats, y_train_combined_feats)
regressor2 = LinearRegression().fit(X_train_keyword_only_feats, y_train_keyword_only_feats)
regressor3 = LinearRegression().fit(X_train_nonkeyword_feats, y_train_nonkeyword_feats)
```

```python
# Predict results
y_pred1 = regressor1.predict(X_test_combined_feats)
y_pred2 = regressor2.predict(X_test_keyword_only_feats)
y_pred3 = regressor3.predict(X_test_nonkeyword_feats)
#print(y_pred)
```

```python
print(regressor1.coef_)
# print(regressor.intercept_)
```

#### Accuracy Calculation

```python
# regressor.score(X_train, y_train)
# regressor.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

print("Prediction on test sets: \n")

# For combined features
print("Combined features: ")
print(mean_squared_error(y_test_combined_feats, y_pred1, squared=False))
print(mean_absolute_error(y_test_combined_feats, y_pred1))
print(mean_absolute_percentage_error(y_test_combined_feats, y_pred1), "\n")

# For keyword only features
print("Keyword only features: ")
print(mean_squared_error(y_test_keyword_only_feats, y_pred2, squared=False))
print(mean_absolute_error(y_test_keyword_only_feats, y_pred2))
print(mean_absolute_percentage_error(y_test_keyword_only_feats, y_pred2), "\n")

# For nonkeyword  features
print("Nonkeyword features: ")
print(mean_squared_error(y_test_nonkeyword_feats, y_pred3, squared=False))
print(mean_absolute_error(y_test_nonkeyword_feats, y_pred3))
print(mean_absolute_percentage_error(y_test_nonkeyword_feats, y_pred3))
```
